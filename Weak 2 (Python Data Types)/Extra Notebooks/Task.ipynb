### **`Problem-1:`** Write a Python function that takes a list and returns a new list with unique elements of the first list.

**Exercise 1:**

Input:

```bash
[1,2,3,3,3,3,4,5]
```

Output:

```bash
[1, 2, 3, 4, 5]
```

````python
def unique(List):
    """
    This function get list of number and return unique numbers
    input:list of numbers
    output:unique elements in the list
    """
    set1=set(List)
    set1=list(set1)
    return set1
L=[1,2,3,3,3,3,4,5]
print(unique(L))
````

### **`Problem-2:`** Write a Python function that accepts a hyphen-separated sequence of words as parameter and returns the words in a hyphen-separated sequence after sorting them alphabetically.

**Example 1:**

Input:
```bash
green-red-yellow-black-white
```

Output:
```bash
black-green-red-white-yellow
```

````python
# Write code here
def sort_fun(l):
    """
    The sort_fun() function accept a string hypen separated values
    and convert into sorted form and return
    input:unsorted hypen text
    output:sorted hypen text
    """
    newtext=l.split('-')
    newtext.sort()
    newtext='-'.join(newtext)
    return newtext

str1="green-red-yellow-black-white"
print(sort_fun(str1))

````

### **`Problem 5:`** Write a Python function to check whether a number is perfect or not.

A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

Example :

```
The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.
```


````python
# Write code here
def is_perfect(num):
    """
    this function accept input number and check number is perfect or not
    input:number (integer)
    output:perfect or not perfect
    """
    l=[]
    for i in range(1,(num//2)+1):
        if num%i==0:
            l.append(i)
    sum=0
    for j in l:
        sum=sum+j
    if (sum+num)/2==num:
        return f"{num} Number is Perfect"
    else:
        return f"{num} Number is not Perfect"

n=496
is_perfect(n)
````

### **`Problem-6:`** Write a Python function to concatenate any no of dictionaries to create a new one.

```
Sample Dictionary :
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}
Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
```

````python
# Write code here
def concatenate_dictionary(list_dict):
    new_dic={}
    for i in list_dict:
        new_dic.update(i)
    return new_dic


dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}
l=[dic1,dic2,dic3]
print(concatenate_dictionary(l))
````

`Problem-8` Write a python function that receives a list of integers and prints out a histogram of bin size 10

```
Input:
[13,42,15,37,22,39,41,50]
```

```
Output:
{11-20:2,21-30:1,31-40:2,41-50:3}
```

````python
# Write code here
def histogram(data):
    # Find the minimum and maximum values in the data
    min_value = min(data)
    max_value = max(data)


    # Create a histogram using a dictionary to count occurrences in each bin
    histogram_dict = {}
    for value in data:
        bin_index = (value - min_value) // 10
        bin_start = min_value + bin_index * 10
        if bin_start not in histogram_dict:
            histogram_dict[bin_start] = 1
        else:
            histogram_dict[bin_start] += 1

    # Print the histogram
    print("Histogram:")
    for bin_start, count in sorted(histogram_dict.items()):
        bin_end = bin_start + 9
        print(f"{bin_start}-{bin_end}: {count}")

# Example usage:
data_list = [5, 15, 25, 35, 45, 12, 18, 22, 32, 38, 52, 5, 15, 28, 38]
histogram(data_list)
````

`Problem-9` Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate which is closest in terms of distance from the query point.

```
List of Coordinates
[(1,1),(2,2),(3,3),(4,4)]
Query Point
(0,0)
```

```
Output
Nearest to (0,0) is (1,1)
```

````python
# Write code here
import math

def find_closest_coordinate(coordinates, query_point):
    if not coordinates:
        return None  # Return None if the list of coordinates is empty

    # Calculate the distance between each coordinate and the query point
    distances = [math.sqrt((x - query_point[0])**2 + (y - query_point[1])**2) for x, y in coordinates]

    # Find the index of the coordinate with the minimum distance
    min_distance_index = distances.index(min(distances))

    # Return the closest coordinate
    return coordinates[min_distance_index]

# Example usage:
coordinate_list = [(1, 1), (2, 2), (3, 3), (4, 4)]
query_point = (0, 0)

closest_coordinate = find_closest_coordinate(coordinate_list, query_point)
print(f"The closest coordinate to {query_point} is: {closest_coordinate}")
````

`Problem-14`: Use reduce to convert a 2D list to 1D

````python
# Write code here
from functools import reduce

# Example 2D list
two_dimensional_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Use reduce to concatenate sublists into a single list
one_dimensional_list = reduce(lambda x, y: x + y, two_dimensional_list)

print("Original 2D List:")
for row in two_dimensional_list:
    print(row)

print("\nConverted 1D List:", one_dimensional_list)
````
